# naive bayes
- p(class|data)=(p(data|class) * p(class)) / p(data)

# probability of playing
- p(play=yes|overcast)=(p(overcast|yes)*p(yes))/p(overcast)==>eqn (1)
- calculate prior probabilities
     - p(overcast)=4/14=0.29
- calculate postrior probabilities
     - p(yes)=9/14=0.64
- put prior and posterior probabilities in eqn(1)
     - p(yes|overcast)=0.44*0.64/0.29=0.98(Higher)

# probability of not playing
- p(no|overcast)=(p(overcast|no)*p(no))/p(overcast)==>eqn (2)
- calculate prior probabilities
     - p(overcast)=4/14=0.29
     - p(no)=5/14=0.36
- calculate postrior probabilities
     - p(overcast|no)=0/9=0
- put prior and posterior probabilities in eqn(2)
     - p(no|overcast)=0*0.36/0.29=0
